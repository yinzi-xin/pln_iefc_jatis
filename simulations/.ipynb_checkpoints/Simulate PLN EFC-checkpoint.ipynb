```python
import numpy as np
import matplotlib.pyplot as plt
from pln_iefc_sims import *
```

```python
### CODE ADAPTED FROM HCIPY TUTORIAL ON EFC: https://docs.hcipy.org/dev/tutorials/ElectricFieldConjugation/ElectricFieldConjugation.html

epsilon = 1e-4
def get_jacobian_matrix(get_image, num_modes, dark_zone = [1,2,3,4], current_actuators = None):
    responses = []
    amps = np.linspace(-epsilon, epsilon, 2)

    if current_actuators is None:
        current_actuators = np.zeros(num_modes)
        
    for i, mode in enumerate(np.eye(num_modes)):
        response = 0

        for amp in amps:
            response += amp * get_image(tt_x=x_center,tt_y=y_center,dm1_acts=mode * amp+current_actuators).electric_field

        response /= np.var(amps)
        response = response[dark_zone]

        responses.append(np.concatenate((response.real, response.imag)))

    jacobian = np.array(responses).T
    return jacobian
```

```python
def run_efc(get_image, num_modes, jacobian, rcond=1e-2, dark_zone = [1,2,3,4], num_iters=50, current_actuators = None):
    # Calculate EFC matrix
    efc_matrix = inverse_tikhonov(jacobian, rcond)

    # Run EFC loop
    if current_actuators is None:
        current_actuators = np.zeros(num_modes)

    actuators = []
    electric_fields = []
    images = []

    for i in range(num_iters):
        img = get_image(tt_x=x_center,tt_y=y_center,dm1_acts=current_actuators)

        electric_field = img.electric_field
        image = img.intensity

        actuators.append(current_actuators.copy())
        electric_fields.append(electric_field)
        images.append(image)

        x = np.concatenate((electric_field[dark_zone].real, electric_field[dark_zone].imag))
        y = efc_matrix.dot(x)

        current_actuators -= efc_loop_gain * y

    return np.array(actuators), np.array(electric_fields), np.array(images)
```

```python
efc_loop_gain = 0.5
jacobian = get_jacobian_matrix(prop_pln_lant2, len(influence_functions))
actuators_1, electric_fields_1, images_1 = run_efc(prop_pln_lant2, len(influence_functions), jacobian, num_iters = 3)
jacobian = get_jacobian_matrix(prop_pln_lant2, len(influence_functions),current_actuators = actuators_1[-1])
actuators_2, electric_fields_2, images_2 = run_efc(prop_pln_lant2, len(influence_functions), jacobian, num_iters = 50, current_actuators = actuators_1[-1])
```

```python
null_ports = ['LP 11a', 'LP 11b', 'LP 21a', 'LP 21b']

plt.figure()
plt.semilogy(np.vstack((images_1[:,1:5],images_2[:,1:5])),label=null_ports)
plt.xlabel('Iteration')
plt.ylabel(r'$\eta_s$')
plt.legend()

last_dm = actuators_2[-1]
outputs,integrands = prop_pln_lant2(tt_x=x_center,tt_y=y_center,dm1_acts=last_dm,return_integrand=True)

plt.figure()
for n in range(4):
    plt.subplot(1,4,n+1)
    imshow_field(integrands[n+1])
plt.show()

plt.figure()
plt.imshow(last_dm.reshape((12,12)))
```
